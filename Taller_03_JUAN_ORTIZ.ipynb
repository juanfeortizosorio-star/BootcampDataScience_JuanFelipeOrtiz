```python
## SE IMPORTAN LAS LIBRERÍAS NECESARIAS PARA REALIZAR LA PETICIÓN A LA API Y MANEJAR LOS DATOS OBTENIDOS DE LA API
import requests
import pandas as pd

base_url = "https://f1api.dev"
endpoint = "/api/circuits"
response = requests.get(base_url + endpoint)

if response.status_code == 200:
  print("Conexión exitosa a la API. Codigo de estado:", response.status_code)
else:
  print("Conexión fallida a la API. Codigo de estado", response.status_code)
```

```python
## SE OBTIENEN LOS DATOS DE LA API EN FORMATO JSON Y SE IMPRIMEN LAS CLAVES DEL DICCIONARIO RESULTANTE
data = response.json()
print(data.keys())
```

```python
## SE IMPRIME EL TOTAL DE CIRCUITOS OBTENIDOS DE LA API
print("Total de circuitos:", len(data['circuits']))
```

```python
## SE CREA UN DATAFRAME DE PANDAS A PARTIR DE LOS DATOS OBTENIDOS DE LA API Y SE MUESTRA LAS PRIMERAS FILAS DEL DATAFRAME
df = pd.DataFrame(data['circuits'])

df.head()
```

```python
##  SE IMPRIMEN LA INFORMACIÓN DEL DATAFRAME PARA VER LOS TIPOS DE DATOS DE CADA COLUMNA Y LA CANTIDAD DE VALORES NO NULOS EN CADA COLUMNA
print(df.info())
```

```python
## SE IMPRIME LA CANTIDAD DE VALORES NULOS EN CADA COLUMNA DEL DATAFRAME
print(df.isnull().sum())
```

```python
## SE ELIMINAN LOS VALORES NULOS DEL DATAFRAME Y SE IMPRIME LA CANTIDAD DE VALORES NULOS EN CADA COLUMNA PARA VERIFICAR QUE SE HAN ELIMINADO LOS VALORES NULOS
df = df.drop_duplicates()
```

```python
## SE ELIMINA LA COLUMNA 'url' DEL DATAFRAME YA QUE NO ES NECESARIA PARA EL ANÁLISIS DE LOS DATOS
df = df.drop(columns=['url'])
```

```python
## SE ELIMINAN LOS VALORES NULOS DE LA COLUMNA 'circuitLength' YA QUE ES UNA COLUMNA IMPORTANTE PARA EL ANÁLISIS DE LOS DATOS
df = df.dropna(subset=['circuitLength'])
```

```python
## SE RENOMBRAN LAS COLUMNAS DEL DATAFRAME PARA QUE LOS NOMBRES DE LAS COLUMNAS SEAN MÁS DESCRIPTIVOS Y FÁCIL DE ENTENDER EN ESPAÑOL
df = df.rename(columns={
    'circuitId': 'id_circuito',
    'circuitName': 'nombre_circuito',
    'country': 'pais',
    'city': 'ciudad',
    'circuitLength': 'longitud_metros',
    'lapRecord': 'record_vuelta',
    'firstParticipationYear': 'primer_YEAR_f1',
    'numberOfCorners': 'numero_curvas',
    'fastestLapDriverId': 'piloto_record',
    'fastestLapTeamId': 'equipo_record',
    'fastestLapYear': 'YEAR_record'
})
```

```python
## SE IMPRIME LA INFORMACION DEL LOS PILOTOS QUE TIENEN EL RECORD DE VUELTA EN LOS CIRCUITOS OBTENIDOS DE LA API
pilotos = df['piloto_record'].value_counts()

print(pilotos)
```

```python
## SE IMPRIME LOS 5 PILOTOS QUE TIENEN EL RECORD DE VUELTA EN LOS CIRCUITOS OBTENIDOS DE LA API
print(df['piloto_record'].value_counts().head(5))
```

```python
## SE IMPRIME LA INFORMACION DE LOS EQUIPOS QUE TIENEN EL RECORD DE VUELTA EN LOS CIRCUITOS OBTENIDOS DE LA API
equipos = df['equipo_record'].value_counts()

print(equipos)
```

```python
## SE CAMBIAN LOS TIPOS DE DATOS DE LAS COLUMNAS 'primer_YEAR_f1', 'YEAR_record', 'numero_curvas' Y 'longitud_metros' A ENTEROS PARA REALIZAR ANÁLISIS DE LOS DATOS MÁS FÁCILMENTE
df['primer_YEAR_f1'] = df['primer_YEAR_f1'].astype(int)
df['YEAR_record'] = df['YEAR_record'].astype(int)
df['numero_curvas'] = df['numero_curvas'].astype(int)
df['longitud_metros'] = df['longitud_metros'].astype(int)
```

```python
## SE GUARDAN LOS DATOS LIMPIOS EN UN NUEVO ARCHIVO CSV PARA REALIZAR ANÁLISIS DE LOS DATOS MÁS FÁCILMENTE
df.to_csv('circuitos_f1_limpio.csv', index=False)
```
